```python
import datetime

today_date = datetime.datetime.today()
```

```python
today_date-datetime.timedelta(1) # <--This function compare the days
```

```python
### List Comprehension

days = [today_date-datetime.timedelta(x) for x in range(0,15)]
```

```python
import pandas as pd
data = pd.DataFrame(days)
data.columns = ["Day"]
data.head()
```

```python
data['weekday'] = data['Day'].dt.day_name() #This function is amazing -dou
data.head()
```

```python
dictionary = {
    
    'Monday':1,
    'Tuesday':2,
    'Wednesday':3,
    'Thursday':4,
    'Friday':5,
    'Saturday':6,
    'Sunday':7
}
```

```python
dictionary
```

```python
data['Weekday_ordinal'] = data['weekday'].map(dictionary)
```

#### Count or Frequency Encoding

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head()
```

```python
columns=[1,3,5,6,7,8,9,13]
train_set = train_set[columns]
```

```python
train_set.columns=['Employment','Degree','Status','Designation','Family_Job','Race','Sex','Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns[:]:
    print(feature,":",len(train_set[feature].unique()),'labels')
```

```python
train_set
```

```python
country_map = train_set['Country'].value_counts().to_dict() #Question Mark is NaN values
                                            #For NaN:
                                        #Apply random sample imputation
                                        #Create another feature "others"
```

```python
train_set['Country']=train_set['Country'].map(country_map)
#Summary: We convert the "country" feature into integer
```

```python
train_set.head(20)
```

#### Advantages

1. Easy to use 
2. Not increasing any feature space

#### Disadvantages

1. It will provide same weight if the frequencies are same but it's very rare


# Target-guided ordinal encoding
1. Ordering the labels according to the target
2. Replace the labels by the joint probability of being 1 or 0

```python
import pandas as pd 
df = pd.read_csv(r'C:\\Users\\doguy\\Desktop\\Documentations\\titanic.csv', usecols=['Cabin','Survived'])
df.head(10)
```

```python
df['Cabin'].fillna('Missing',inplace=True)
df.head()
```

```python
df['Cabin']= df['Cabin'].astype(str).str[0]
```

```python
df.groupby([df.Cabin])['Survived'].mean()
```

```python
ordinal_labels =  df.groupby([df.Cabin])['Survived'].mean().sort_values().index

ordinal_labels
```

```python
ordinal_labels2 = {k:i for i,k in enumerate(ordinal_labels,0)} 
```

```python
ordinal_labels2
```

```python
df['Cabin_ordinal'] = df['Cabin'].map(ordinal_labels2)
df.head()
```
